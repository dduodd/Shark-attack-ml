#extract text from pdf
#(takes around 1h 30 min to run)

```python
import requests
import pdfplumber
import pandas as pd
import io

# Load your existing DataFrame with the PDF links
# Assuming your DataFrame is already loaded and named df

# Initialize a list to store extracted text
text_data = []

# Iterate through the DataFrame and process PDFs
for index, row in df.iterrows():
    url = row['href_formula']
    try:
        response = requests.get(url)

        if response.status_code == 200:
            with pdfplumber.open(io.BytesIO(response.content)) as pdf:
                pdf_text = ""
                for page in pdf.pages:
                    pdf_text += page.extract_text()

            # Append the extracted text to the list
            text_data.append(pdf_text)
        else:
            text_data.append("")  # Handle the case when a PDF can't be retrieved
    except requests.exceptions.RequestException:
        # Handle any exceptions, such as a 404 error
        text_data.append("")  # Here, you can choose to add a specific message or leave it blank

# Now, you have text_data containing text extracted from PDFs (including any exceptions)
# You can proceed with the topic modeling code as previously described
```

```python
pdftext = {'Text': text_data}
df_pdftext = pd.DataFrame(pdftext)    #put text in df called df_pdftext
```

```python
#remove puncutation
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')

# Define a function to remove punctuation from a text
def remove_punctuation_with_nltk(text):
    # Tokenize the text using NLTK
    tokens = word_tokenize(text)

    # Filter out words that are not punctuation
    tokens = [word for word in tokens if word.isalpha()]

    # Join the tokens back into a single string
    cleaned_text = ' '.join(tokens)

    return cleaned_text

# Apply the function to your text column (e.g., 's7_rejoin')
df_pdftext['s5_punctuation'] = df_pdftext['s4_narrative'].apply(remove_punctuation_with_nltk)
```

```python
#tokenize
import nltk
nltk.download('punkt')

from nltk.tokenize import word_tokenize
df_pdftext['s6_tokenize'] = df_pdftext['s5_punctuation'].apply(lambda x: word_tokenize(x))
```

```python
#remove stopword
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

def remove_stopwords(tokens):
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
    return filtered_tokens

df_pdftext['s7_stopword'] = df_pdftext['s6_tokenize'].apply(lambda tokens: remove_stopwords(tokens))
```

```python
#lemmatize/stemming
from nltk.stem import WordNetLemmatizer
nltk.download ('wordnet')

lemmatizer = WordNetLemmatizer()

def lemmatize_text(tokens):
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return lemmatized_tokens

df_pdftext['s8_lemmatized'] = df_pdftext['s7_stopword'].apply(lemmatize_text)
```

```python
# Join the tokenized words into a single string for each row
df_pdftext['s9_rejoin'] = df_pdftext['s8_lemmatized'].apply(lambda tokens: ' '.join(tokens))
```

```python
#rid of null value
df_pdftext = df_pdftext.dropna()
```

```python
#tf-idf vectorize

from sklearn.feature_extraction.text import TfidfVectorizer

# TfidfVectorizer
vectorizer = TfidfVectorizer(max_features=10000, stop_words='english')  # You can adjust max_features as needed

# Fit vectorizer
tfidf_matrix = vectorizer.fit_transform(df_pdftext['s9_rejoin'])

# Convert the TF-IDF matrix to a DataFrame for easier manipulation (optional)
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())
tfidf_df
```

```python
#kmean - 1st run

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import pandas as pd

# Apply to K-Mean
kmeans = KMeans(n_clusters=7, random_state=42)
kmeans.fit(tfidf_matrix)

# Assign cluster to df
df_pdftext['cluster_label'] = kmeans.labels_

# Check the distribution of documents across clusters
print(df_pdftext['cluster_label'].value_counts())
```

```python
#drop all record associated with 0 and 4 (useless reports)
df_pdftext = df_pdftext[~df_pdftext['cluster_label'].isin([0, 4])]
```

```python
#kmean - 2nd run

# TfidfVectorizer
vectorizer = TfidfVectorizer(max_features=10000, stop_words='english')

# Fit TfidfVectorizer and apply to K-Mean
tfidf_matrix2 = vectorizer.fit_transform(df_pdftext['cluster_label'])
kmeans = KMeans(n_clusters=7, random_state=42)
kmeans.fit(tfidf_matrix2)

# Cluster to df
df_pdftext['cluster_label2'] = kmeans.labels_

# Count # of cluster
print(df_pdftext['cluster_label2'].value_counts())
```

```python
#drop all record associated with 0 and 4 (useless reports)
df_pdftext = df_pdftext[~df_pdftext['cluster_label2'].isin([0])]
```

```python
# Assuming 'cluster_label' is the column with the initial clusters
unique_clusters = df_pdftext['cluster2'].unique()

# Function to recluster within each existing cluster
def recluster_within_cluster(cluster_df):
    vectorizer = TfidfVectorizer(max_features=10000, stop_words='english')

    # Fit TfidfVectorizer and apply to K-Mean
    tfidf_matrix = vectorizer.fit_transform(cluster_df['s9_rejoin'])

    # Use the same number of clusters as the original cluster
    kmeans = KMeans(n_clusters=len(cluster_df['cluster_label'].unique()), random_state=42)
    kmeans.fit(tfidf_matrix)

    # Assign new cluster labels within the existing cluster
    cluster_df['cluster3'] = kmeans.labels_

    return cluster_df

# Apply recluster_within_cluster function to each unique cluster
df_pdftext = df_pdftext.groupby('cluster_label').apply(recluster_within_cluster).reset_index(drop=True)
```

```python
#kmean - 3rd run

# TfidfVectorizer
vectorizer = TfidfVectorizer(max_features=10000, stop_words='english')

# Fit TfidfVectorizer and apply to K-Mean
tfidf_matrix3 = vectorizer.fit_transform(df_pdftext['s9_rejoin'])
kmeans = KMeans(n_clusters=7, random_state=42)
kmeans.fit(tfidf_matrix3)

# Cluster to df
df_pdftext['cluster_label3'] = kmeans.labels_

# Count # of cluster
print(df_pdftext['cluster_label3'].value_counts())
```

```python
from sklearn.metrics import silhouette_score

# Assessing model accuracy using silhouette score
silhouette_avg = silhouette_score(tfidf_matrix3, kmeans.labels_)
print(f"K-Means clustering score: {silhouette_avg}")
```
